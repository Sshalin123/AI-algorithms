Spline Interpolation
Spline interpolation connects a series of data points with piecewise polynomial functions, ensuring a smooth transition between the pieces.

How it works: Instead of one high-degree polynomial, it uses multiple low-degree polynomials (like cubic splines) to connect subsets of points.

Best for: Creating very smooth curves for large datasets without the oscillation problems of single polynomial interpolation. It's widely used in computer graphics, data visualization, and engineering


```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# 1. Create a 2D grid of known points (5x5)
y_known_2d = torch.randn(5, 5)
# Unsqueeze to add Batch and Channel dimensions: (1, 1, 5, 5)
y_known_4d = y_known_2d.view(1, 1, 5, 5)

# 2. Perform bicubic interpolation to a 100x100 grid
y_interp_4d = F.interpolate(y_known_4d, size=(100, 100), mode='bicubic', align_corners=True)

# Squeeze back to 2D
y_interp = y_interp_4d.squeeze()

# 3. Plotting
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(y_known_2d, origin='lower', extent=(0, 1, 0, 1))
plt.title("Known Points (5x5)")
plt.subplot(1, 2, 2)
plt.imshow(y_interp, origin='lower', extent=(0, 1, 0, 1))
plt.title("Bicubic Spline Interp (100x100)")
plt.show()
```
