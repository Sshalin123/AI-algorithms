Linear Interpolation
This method assumes a straight line connects two known data points and estimates the value based on that line.

How it works: Calculates a weighted average of the two nearest data points.

Best for: Datasets where the points are expected to have a roughly linear relationship. It's a good balance between simplicity and accuracy for many applications.

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# 1. Known data points
y_known = torch.tensor([0., 0.8, 0.9, 0.1, -0.8])
# Unsqueeze to add Batch and Channel dimensions: (1, 1, 5)
y_known_3d = y_known.view(1, 1, -1)

# 2. Perform interpolation
y_interp_3d = F.interpolate(y_known_3d, size=100, mode='linear', align_corners=True)

# Squeeze back to 1D
y_interp = y_interp_3d.squeeze()

# 3. Plotting
x_known_coords = torch.linspace(0, 1, len(y_known))
x_interp_coords = torch.linspace(0, 1, len(y_interp))
plt.figure(figsize=(8, 5))
plt.plot(x_known_coords, y_known, 'ro', label='Known Points')
plt.plot(x_interp_coords, y_interp, 'g-', label='Linear Interp')
plt.title("Linear Interpolation")
plt.legend()
plt.show()
```
